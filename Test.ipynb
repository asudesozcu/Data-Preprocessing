```python
from data_preprocessing_library_sevvalcucuk_asudesozcu.MissingValueHandler import MissingValueHandler 
from IPython.display import display

file_path = r"C:\Users\sozcu\Downloads\synthetic_sample_data (1).csv"
handler = MissingValueHandler(file_path)
original_data = handler.data
print("Original Data:")
display(original_data)
#print(original_data)

print("Mean:")
mean_data = handler.impute_mean(original_data)
handler.print_modified_rows(original_data, mean_data)

print("Median:")
median_data = handler.impute_median(original_data)
handler.print_modified_rows(original_data, median_data)

print("Mode:")
mode_data = handler.impute_mode(original_data)
handler.print_modified_rows(original_data, mode_data)

print("Constant:")
constant_data = handler.impute_constant(original_data, 777)
handler.print_modified_rows(original_data, constant_data)

print("Forward Fill:")
ffill_data = handler.impute_ffill(original_data)
handler.print_modified_rows(original_data, ffill_data)

print("Backward Fill :")
bfill_data = handler.impute_bfill(original_data)
handler.print_modified_rows(original_data, bfill_data)

print("Data after Dropping Rows with Missing Values:")
drop_rows_df = handler.drop_missing_rows(handler.data)
display(drop_rows_df)


print("Data after Dropping Columns with Missing Values:")
drop_columns_df = handler.drop_missing_columns(handler.data)
display(drop_columns_df)
```

```python
from data_preprocessing_library_sevvalcucuk_asudesozcu.FeatureEngineer import FeatureEngineer

feature=FeatureEngineer()
df = pd.read_csv(r"C:\Users\sozcu\Desktop\şevvalin_datası.csv")
print("Original DataFrame:\n")
display(df)
print("\n")

print("Dataframe Sorted Alphabetically According To Movie ID Column (A-Z):\n")
abc_df = feature.sort_df_alphabetically(df,'Movie Id',True)
display(abc_df)
print("\n")

print("Dataframe Sorted Alphabetically According To Movie ID Column (Z-A):\n")
abc_df = feature.sort_df_alphabetically(df,'Movie Id', False)
display(abc_df)
print("\n")

print("Dataframe Sorted By Date According To Release Date Column (Old to New):\n")
date_df = feature.sort_df_by_date(df, 'Release Date')
display(date_df)
print("\n")

print("Dataframe Sorted By Date According To Release Date Column (New to Old):\n")
date_df = feature.sort_df_by_date(df, 'Release Date', False)
display(date_df)
print("\n")

print("Dataframe Filtered By Key (Horror) According To Genre Column (Only Horrors):\n")
key_df = feature.filter_df_by_key(df, 'Genre' , "Horror")
display(key_df)
print("\n")

print("Dataframe Filtered By Key (Horror) According To Genre Column (Excluding Horror):\n")
key_df = feature.filter_df_by_key(df, 'Genre' , "Horror", False)
display(key_df)
print("\n")

print("Dataframe Sorted By Key (7) According To Rating Column (Bigger Than 7):\n")
key_df = feature.sort_df_by_key(df, 'Rating' , 7)
display(key_df)
print("\n")

print("Dataframe Sorted By Key (7) According To Rating Column (Smaller Than 7):\n")
key_df = feature.sort_df_by_key(df, 'Rating' , 7, False)
display(key_df)
print("\n")


```

```python
from data_preprocessing_library_sevvalcucuk_asudesozcu.DateTimeHandler import DateTimeHandler
from datetime import datetime
handler= DateTimeHandler()
df = pd.read_csv(r"C:\Users\sozcu\Desktop\şevvalin_datası.csv")
print("Original DataFrame:\n")
display(df)
print("\n")

print("\nFiltered DataFrame (between 1990-01-01 and 2000-12-31):")
filtered_df = handler.filter_dates(df, ['Release Date'], criteria='min_max', min_date=datetime(1990, 1, 1), max_date=datetime(2000, 12, 31))
display(filtered_df)
print("\n")

print("\nFiltered DataFrame (between 09:00:00 and 17:00:00):")
filtered_df = handler.filter_time(df, ['Release Time'], criteria='min_max', min_time=datetime.strptime('09:00:00', '%H:%M:%S'), max_time=datetime.strptime('15:00:00', '%H:%M:%S'))
display(filtered_df)
print("\n")

print("\nDates Extracted DataFrame :")
date_extracted_df = handler.extract_date_parts(df, ['Release Date'])
display(date_extracted_df)
print("\n")

print("\nTime Extracted DataFrame :")
time_extracted_df = handler.extract_time_parts(df, ['Release Time'])
display(time_extracted_df)
print("\n")

print("\nDate Replaced (from 7/7/1985 to 21/5/2024) DataFrame:")
df_replaced_date = handler.replace_date(df, ['Release Date'], '7/7/1985', datetime(2024, 5, 21))
print("\nDataFrame with replaced date:")
display(df_replaced_date)

print("\nTime Replaced (from 3:00 PM to 1:30 PM) DataFrame:")
df_replaced_date = handler.replace_time(df, ['Release Time'], '3:00 PM', '1:30 PM')
print("\nDataFrame with replaced date:")
display(df_replaced_date)
```

```python
from data_preprocessing_library_sevvalcucuk_asudesozcu.DataTypeConverter import DataTypeConverter 

converter = DataTypeConverter()

df = pd.read_csv(r"C:\Users\sozcu\Desktop\şevvalin_datası.csv")
print("Original DataFrame:\n")
display(df)
print("\n")

print("Converting Rating Column To Numeric...\n")

df['Rating'] = df['Rating'].astype('object')
print("Before conversion:")
print("Type of 'Rating' column:", df['Rating'].dtype)
print()

df = converter.to_numeric(df, ['Rating'])

display(df)

print("\nAfter conversion:")
print("Type of 'Rating' column:", df['Rating'].dtype)
print("\n")


print("Converting Genre Column To Numeric...\n")

print("Before conversion:")
print("Type of 'Genre' column:", df['Genre'].dtype)
print()

df = converter.to_numeric(df, ['Genre'])

display(df)

print("\nAfter conversion:")
print("Type of 'Genre' column:", df['Genre'].dtype)
print("\n")



print("Converting Movie Id Column To Categorical...\n")

print("Before conversion:")
print("Type of 'Movie Id' column:", df['Movie Id'].dtype)
print()

df = converter.to_categorical(df, ['Movie Id'])

display(df)

print("\nAfter conversion:")
print("Type of 'Movie Id' column:", df['Movie Id'].dtype)
print()
```

```python
from data_preprocessing_library_sevvalcucuk_asudesozcu.CategoricalEncoder import CategoricalEncoder 
import pandas as pd
encoder = CategoricalEncoder()
df = pd.read_csv(r"C:\Users\sozcu\Desktop\şevvalin_datası.csv")
print("Original DataFrame:\n")
display(df)
print("\n")

print("One Hot Encoded Dataframe:\n")
one_hot_df = encoder.one_hot_encode(df, ['Genre'])
display(one_hot_df)
print("\n")

print("Inversed One Hot Encoded Dataframe:\n")
one_hot_df = encoder.inverse_one_hot_encode(one_hot_df, ['Genre'])
display(one_hot_df)
print("\n")

print("Label Encoded Dataframe:\n")
label_df = encoder.label_encode(df, ['Genre'])
display(label_df)
print("\n")

print("Inversed Label Encoded Dataframe:\n")
label_df = encoder.inverse_label_encode(label_df, ['Genre'])
display(label_df)
print("\n")



print("Frequency Encoded Dataframe:\n")
frequency_df = encoder.frequency_encode(df, ['Genre'])
display(frequency_df)
print("\n")

print("Binary Encoded Dataframe:\n")
binary_df = encoder.binary_encode(df, ['Popular'])
display(binary_df)
print("\n")
```
